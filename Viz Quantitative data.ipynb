```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
listings = pd.read_csv('AB_NYC_2019.csv')
```

```python
listings
```

```python
plt.hist(listings['price'])
plt.ylabel('host_name')
plt.xlabel('price(in US Dollar)')
plt.show()
```

```python
plt.hist(listings['price'], bins= np.arrange(0, 110, 40))
plt.ylabel('host_name')
plt.xlabel('price(in US Dollar)')
plt.show()
```

```python
plt.hist(listings['price'], bins= np.arange(0, 110, 5))
plt.ylabel('host_name')
plt.xlabel('price(in US Dollar)')
plt.show()
```

```python
plt.scatter(listings['price'], y =listings['number_of_reviews'])
plt.ylabel('numbers of reviews')
plt.xlabel('price(in US Dollar)')
plt.show()
```

```python
plt.scatter(listings['price'], y =listings['number_of_reviews'], s=8)
plt.ylabel('numbers of reviews')
plt.xlabel('price(in US Dollar)')
plt.xlim(0, 1100)
plt.show()
```
